```python
# test resAdjTransformer
from tpot.builtins import resAdjTransformer
C_list = ['pc1', 
         'pc2', 
         'pc3', 
         'pc4', 
         'pc5', 
         'pc6', 
         'pc7', 
         'pc8', 
         'pc9', 
         'pc10']
trans = resAdjTransformer(C=C_list, adj_list='testingData/basicTest/adj_list.csv')
trans.fit(X, y)
X_adj = trans.transform(X)
```

```python
# test all preprocessors can ignore adjY and indicators
# test it in a pipeline
from sklearn.pipeline import make_pipeline
reg = resAdjDecisionTreeRegressor(random_state=42)
for file in glob('tpot/builtins/resAdj_preprocessors/res*py'):
    
    opname = file.split('/')[-1].split('.')[0]
    print('test selection with', opname)
    sel = eval(opname)()
    if hasattr(sel, 'random_state'):
        sel.set_params(random_state=42)
    print(sel)
    est = make_pipeline(sel, reg)
    cv_scores_1 = cross_val_score(est, X, y.values, scoring=resAdjMseScorer, cv=resAdjPredefinedSplits(n_splits=5), error_score='raise')
    cv_scores_2 = cross_val_score(est, X, X['array'].values, scoring=resAdjMseScorer, cv=resAdjPredefinedSplits(n_splits=5))
    if np.array_equal(cv_scores_1, cv_scores_2):
        print('{} passed'.format(opname))
    else:
        print('{} failed'.format(opname))
        break
```
